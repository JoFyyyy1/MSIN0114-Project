### BRAND: Implement missing values from the known infor  [activity_log]
the pattern of 'activity_log': item_id, cat_id, brand_id, time_stamp

```python
missing_brand = df_ud[(df_ud['brand_id'].isnull() == True) &(df_ud['activity'].isnull() == False)]
missing_brand.isnull().sum()/missing_brand.shape[0]
```

```python
missing_brand.shape
```

```python
missing_brand
```

```python
def insert_brand(item,activity):
    try:
        for i in activity: #missing_brand.iloc[1,15]
            if int(i[0]) == item:
                return int(i[2])
    except TypeError:
        return np.nan
```

```python
missing_brand['brand'] = missing_brand.apply(lambda x: insert_brand(x.item_id,x.activity),axis = 1)
missing_brand['brand'].isnull().sum()
```

```python
del missing_brand['brand_id']
gc.collect()
# missing_brand.rename(columns = {'brand':'brand_id'},inplace = True)
missing_brand
```

```python
found_index = missing_brand.index[pd.isnull(missing_brand['brand']) == False].tolist()
found_brand = missing_brand.loc[found_index,'brand'].tolist()
brand = df_ud.brand_id.tolist()
for i in  range(len(found_index)):
    brand[found_index[i]] = found_brand[i]
df_ud['brand_id'] = brand
```

### BRAND: Impute by 'item_id'

```python
grouped = df_ud['brand_id'].groupby(df_ud['item_id']) # the brand distribution according of item
groupedl = list(grouped)
item_list = []
brand_mode = []
all_miss = []
for i in range(len(groupedl)):
    if all(np.isnan(groupedl[i][1].values)):
        all_miss.append(groupedl[i][0])
    else:
        # print(i)
        item_list.append(groupedl[i][0])
        brand_mode.append(groupedl[i][1].mode().values[0])
```

```python
item_brand = pd.DataFrame([item_list,brand_mode]).T
item_brand.columns =['item','brand']
item_brand
```

```python
missing_brand = df_ud[df_ud['brand_id'].isnull() == True]
missing_brand
```

```python
item_brand[item_brand['item'] == 692695]
```

```python
def insert_miss_brand(item,brand): # impute the missing value of 'brand'
    if len(item_brand[item_brand['item'] == item].index) != 0:
        loc = item_list.index(item)
        return brand_mode[loc]
    else:
        return brand
```

```python
missing_brand['brand'] = missing_brand.apply(lambda x: insert_miss_brand(x.item_id,x.brand_id),axis = 1)
missing_brand['brand'].isnull().sum()
```

```python
missing_brand
```

```python
found_index = missing_brand.index[pd.isnull(missing_brand['brand']) == False].tolist()
found_brand = missing_brand.loc[found_index,'brand'].tolist()
brand = df_ud.brand_id.tolist()
for i in  range(len(found_index)):
    brand[found_index[i]] = found_brand[i]
df_ud['brand_id'] = brand
```

```python
df_ud.isnull().sum()
```

```python
brand_miss = df_ud[df_ud['item_id'].isin(all_miss)]
brand_miss.groupby('action_type').count()
# the last brand-missing item shares a large part of positive sample, therefroe, it's irrational to delete them.
```

```python
df_ud['brand_id'].replace(np.nan,df_ud.brand_id.max()+1, inplace = True)
# replace with a new brand_id, setting all of them as a new brand type.
```
